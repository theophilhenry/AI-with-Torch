```python
import torch

x = torch.rand(3, requires_grad=True)
print(x)

y = x + 2
print(y)

z = y * y * 2
z = z.mean()
print(z)

z.backward()
print(x.grad)
```

```python
import torch

x = torch.rand(3, requires_grad=True)
with torch.no_grad():
  y = x + 2
  print(y)
z = x + 3
print(z)
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(1):
  model_output = (weights*3).sum()
  print(model_output)
  model_output.backward()
  print(weights.grad)
  weights.grad.zero_() # reset the gradient
```

```python
weights = torch.ones(4, requires_grad=True)
optimizer = torch.optim.SGD(weights, lr=0.01)
optimizer.step()
optimizer.zero_grad()
```

```python
import torch

# Simple backpropagation
weights = torch.tensor(1.0, requires_grad=True)
x = torch.tensor(5.0)
y = torch.tensor(100.0)

for epoch in range(1000):
  y_hat = (weights*x)
  loss = (y_hat-y)**2
  
  loss.backward()
  
  # Update the weights
  with torch.no_grad():
    weights.data = weights.data - 0.01 * weights.grad
    # 0.01 is the learning rate

  weights.grad.zero_()

print(weights)
```
