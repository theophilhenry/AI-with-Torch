```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
bc = datasets.load_breast_cancer()
X, y = bc.data, bc.target
n_samples, n_features = X.shape # (569, 30)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

# Scale data
# Make our feature to have 0 mean and units variance
# Important for logistic regressions
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

# Convert to pytorch and reshape
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
# Reshape the y to get multi-row 1 column data
y_train = y_train.view(y_train.shape[0], -1)
y_test = y_test.view(y_test.shape[0], -1)

# Setup Model
class LogisticRegression(nn.Module):
  def __init__(self, input_dim):
    super(LogisticRegression, self).__init__()
    self.lin1 = nn.Linear(input_dim, 1)
    
  def forward(self, x):
    x = self.lin1(x)
    y_pred = torch.sigmoid(x) # Return value between 0 and 1
    return y_pred
  
model = LogisticRegression(n_features)

# Loss and Optimizer
lr = 0.01
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

num_epoch = 1000
for epoch in range(num_epoch):
  y_pred = model(X_train)
  loss = criterion(y_pred, y_train)
  
  # Backpropagation
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  
  # Show steps
  if (epoch + 1) % 100 == 0:
    print(f'epoch: {epoch + 1}, loss: {loss.item():.4f}')

with torch.no_grad():
  y_pred = model(X_test)
  y_pred_cls = y_pred.round()
  acc = y_pred_cls.eq(y_test).sum() / float(y_test.shape[0])
  print(f'Accuracy = {acc:.4f}')
```
