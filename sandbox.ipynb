```python
class Simple:
  def __init__(self):
    self.message = "Good Morning"

class Client:
  def __init__(self):
    self.class_groups = ['P', 'R', 'S']
    self.trackers = {
        cls: Simple() for cls in self.class_groups
    }

client = Client()
print(client.trackers)
```
