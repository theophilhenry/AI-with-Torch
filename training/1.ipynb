```python
import torch
import torch.nn as nn
from sklearn import datasets

X_tuple, Y_tuple = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
X = torch.tensor(X_tuple, dtype=torch.float32)
Y = torch.tensor(Y_tuple, dtype=torch.float32)
Y = Y.reshape(-1, 1)
print(Y)

class LinearRegression(nn.Module):
  def __init__(self, input_size):
    super(LinearRegression, self).__init__()
    self.fc1 = nn.Linear(input_size, 1)
  def forward(self, x):
    return self.fc1(x)

model = LinearRegression(input_size=1)

criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for i in range(1000):
  prediction = model(X)
  loss = criterion(Y, prediction)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
loss.item()
```

```python
import matplotlib.pyplot as plt

predicted = model(X).detach().numpy()
plt.plot(X_tuple, Y_tuple, 'ro')
plt.plot(X_tuple, predicted, 'b')
plt.show()
```
